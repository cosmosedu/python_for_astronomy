# Approximate luminosity distance integral:

```python
import numpy as np

z = np.linspace(0,1,1000)
integrand = 1/ np.sqrt(0.3*(1+z)**3 + 0.7)
integral = np.trapezoid(integrand,z)

print ("cosmological integral approximation:", integral)
```

# basic error propagation

```python
v = 7000
H0 = 70 
distance = v/H0
print("distance:",distance)
H0_error = 5 

distance_error = distance *(H0_error/H0)
print("distance uncertainity:", distance_error)

```

# Vectorized error propagation

```python
import numpy as np

velocities = np.array([5000,6000,7000])
H0 = 70 
H0_error = 5 
distance = velocities / H0
distance_error = distance *(H0_error / H0)

print ( distance )
print(distance_error)
```
