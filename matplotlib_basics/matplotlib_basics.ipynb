```python
import numpy as np
import matplotlib.pyplot as plt
```

# simple line plot

```python
x = np.linspace(0,10,100)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("x squared")
plt.title("simple quadratic function")

plt.show()
```

# Add grid and legend

```python
plt.plot(x,y, label="y=x^2")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Growth")
plt.legend()
plt.grid(True)

plt.show()
```

# change line style 

```python
plt.plot(x,y,color="red",linestyle="--",linewidth=1,label="y=x^2")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

# scatter plot

```python
z = np.array([0.01,0.02,0.03,0.04])
velocity = 3e5*z

plt.scatter(z,velocity)

plt.xlabel("Redshift(z)")
plt.ylabel("Velocity(km/s)")
plt.title("Redshift vs Velocity")

plt.grid (True)
plt.show
```

```python
y1 = x**2
y2 = x**3

plt.plot(x, y1, label="x^2")
plt.plot(x, y2, label="x^3")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure (figsize=(6,4))

plt.plot(x,y,label="x^2")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label="x^2")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.savefig("quadratic_plot.png",dpi=300)
plt.show()
```
