```python
#example data
magnitudes = [-18.9,-19.5,7,8.6,-6.5]

#print only bright supernovae(M<7)
for m in magnitudes:
    if m<7:
        print("Bright SN:",m)
```

```python
redshift = [0.01,0.02,0.6,0.03]

for z in redshift:
    if z> 0.01 and z<0.6:
        print("valid cosmological SN:",z)
        
```

```python
for m in magnitudes:
    if m < -19:
        print(m, "is bright")
    else:
        print(m, "is normal")
```

```python
bright_sn = [m for m in magnitudes if m < -19]
print(bright_sn)
```
