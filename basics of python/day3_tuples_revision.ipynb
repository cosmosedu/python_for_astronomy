# Tuples
tuple is immutable however list are mutable

```python
coordinates = (10.5, -5.7)
star_info = ("sirius", 8.6,"A-type")

#access elements 
print(coordinates[0]) #RA
print(coordinates[1]) #dec

print(star_info[0]) #name
print(star_info[1]) #distance

#unpacking tuple 
ra,dec = coordinates 
print("RA=",ra)
print("dec=",dec)

name, distance,stype = star_info
print(name,"is",distance,"light- year away")

#loop through tuple 
for item in star_info:
    print(item)

#tuple length
print(len(star_info))

#tuple inside lists 
stars = [("sun",0),
         ("sirius",8.6),
         ("vega",25)
        ]

for star in stars:
    print(star[0],"distance=",star[1])
```
