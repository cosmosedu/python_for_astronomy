```python
# 1. List create
numbers = [1, 2, 3, 4, 5]
stars = ["Sun", "Sirius", "Vega", "Betelgeuse"]

# 2. Access elements
print(numbers[0])      # first element
print(numbers[-1])     # last element
print(stars[1:3])      # slicing

# 3. Modify list
numbers.append(6)
numbers.remove(3)
print(numbers)

# 4. Length of list
print(len(stars))

# 5. Loop through list
for star in stars:
    print("Star:", star)

# 6. Condition inside loop
for star in stars:
    if len(star) > 4:
        print(star, "has more than 4 letters")

# 7. List comprehension (important)
lengths = [len(star) for star in stars]
print(lengths)

# 8. List comprehension with condition
long_stars = [star for star in stars if len(star) > 4]
print(long_stars)
```
