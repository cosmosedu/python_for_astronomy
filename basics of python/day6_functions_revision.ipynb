```python
#basic function
def square(x):
    return x**2

print(square(4))

#function with multiple arguments
def distance_modulus(m, M):
    return m - M

print(distance_modulus(12.5, -19.3))

#function with condition inside 
def classify_supernova(M):
    if M < -19:
        return "Bright Type Ia"
    else:
        return "Normal SN"

print(classify_supernova(-19.5))

#function returning multiple values
def mean_and_std(data):
    mean = sum(data)/len(data)
    variance = sum((x-mean)**2 for x in data)/ len(data)
    std = variance**0.5
    return mean, std

numbers = (3,5,6,9,85,93)

mean, std = mean_and_std(numbers)
print("Mean:",mean)
print("std:",std)

    


```
