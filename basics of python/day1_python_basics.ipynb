# calcuate the velocity at which earth is moving along its orbit around the sun


```python
radius = 1.496e8 #orbital radius in km
period = 3.156e7 #orbital period in s 

#calcualte orbital velocity 
velocity = 2*3.14159*radius/period
print(velocity)
print("orbital velocity=",velocity,"km/s")
print("orbital velocity ={:5.2f}km/s".format(velocity))
```

The method format () inserts the value of the variable in parentheses into the preceding string .The place holder {:5.2f} controls where and in which format the value of the varible "velocity" is insterted.
