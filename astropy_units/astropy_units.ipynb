```python
import numpy as np 
import astropy.units as u 
import astropy.constants as const
```

```python
distance = 10 * u.Mpc
print(distance)
```

```python
velocity = 7000* u.km/u.s
print(velocity)
```

```python
print(distance.to(u.km))
print(velocity.to(u.m/u.s))
```

```python
H0 = 70 * u.km/(u.s*u.Mpc)

distance = 50* u.Mpc

velocity = H0 * distance
print(velocity)
```

```python
print(velocity.to(u.m/u.s))
```

```python
print(const.c)
print(const.G)
```

```python
z = 0.03
velocity = z*const.c
print(velocity.to(u.km/u.s))
```

```python
#simple scitific example
H0 = 70 * u.km / (u.s * u.Mpc)

age = 1 / H0
print(age)
```

```python
print(age.to(u.yr))
```

```python
distances = np.array([10, 20, 30]) * u.Mpc

velocities = H0 * distances
print(velocities)
```
