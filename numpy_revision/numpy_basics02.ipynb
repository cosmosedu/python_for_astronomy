```python
import numpy as np
```

```python
matrix = np.array([[1, 2],
                   [3, 4]])

print(matrix)
print("Shape:", matrix.shape)
print(matrix[0, 1])

#vectorized math
x = np.array([1,2,3,4,5,6])

print(x + 7)
print(x * 5)
print(x**2)
print(np.sqrt(x))
print(np.mean(x))
print(np.std(x))

#broadcasting

x = np.array([1, 2, 3])
y = 10

print(x + y)

#2d broadcasting
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print(A + 10)

#astronomy example

redshift = np.array([0.01,0.02,0.03,0.04])
c = 3e5 #km/s

velocity = c * redshift
print("velocity:",velocity)




```
