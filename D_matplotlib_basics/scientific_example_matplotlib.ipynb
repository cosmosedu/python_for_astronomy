```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Hubble law as an example

```python
h0 = 70 
distance = np.linspace(0,200,100)
velocity = h0 * distance

plt.figure(figsize=(8,6))
plt.plot(distance, velocity, label="Hubble Law")

plt.xlabel("Distance (Mpc)")
plt.ylabel("Velocity (km/s)")
plt.title("Hubble's Law")
plt.legend()
plt.grid(True)

plt.savefig("hubble_law.png",dpi=300)
plt.show()

```
